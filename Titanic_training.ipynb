```python
import seaborn as sns
import numpy as np
import pandas as pd
from scipy import stats as st
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
df=pd.read_csv("titanic-training-data.csv")
```

```python
df.info()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df["Age"].value_counts().plot(kind='barh',color="green")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("freqency")
plt.show()
```

```python
sns.distplot(df["Age"],color="olive")
plt.title("Distribution of Age")
plt.show()
```

```python
sns.violinplot(x="Sex",y="Age",color="green",data=df)
plt.title("Distribution of Age")
plt.show()
```

```python
sns.swarmplot(x="Sex",y="Age",color="green",data=df)
plt.title("Distribution of Age")
plt.show()
```
